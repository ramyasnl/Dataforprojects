```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/ramyasnl/Submission_Projects/main/movie_scores.csv")
df.head()
```

```python
df.info()
```

```python
# List all the columns in the table
df.columns
```

```python
# We only want IMDb data, so create a new table that takes the Film and all the columns relating to IMDB
dfmovie = df[["FILM", "IMDB", "IMDB_norm",
                            "IMDB_norm_round", "IMDB_user_vote_count"]]
dfmovie.head()
```

# Calculating the Average Rating of the Movies

```python
meanrating_df = dfmovie['IMDB'].mean()
print("Average of Ratings is",meanrating_df)
```

# Sorted in Descending Order movies ratings


```python
newdf =dfmovie.sort_values(by=['IMDB'], ascending=[False])
newdf.head()
```
